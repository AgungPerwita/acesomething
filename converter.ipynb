```python
text = open("list.txt", "r", encoding='utf-8')
lines = text.read()
lines = lines.split("\n\n")
```

```python
questions = []
options = [[] for _ in range(len(lines))]

data = {}

for i in range(len(lines)):
    temp = lines[i].split("\n")
    
    n = 0
    tempquestion = ""
    while not temp[n].startswith("    a. ") and not temp[n].startswith("    b. ") and not temp[n].startswith("    c. ") and not temp[n].startswith("    d. "):
        if temp[n][0].isdigit():
            temp2 = temp[n].split(". ",1)
            tempquestion += temp2[1]
        else:
            tempquestion += "\n"
            tempquestion += temp[n]
        n += 1
    questions.append(tempquestion)

    tempoptions = ""
    if (temp[n][:7] == "    a. "):
        tempoptions += temp[n][7:]
        n = n + 1
    while not temp[n].startswith("    b. "):
        tempoptions += "\n"
        tempoptions += temp[n][7:]
        n = n + 1
    options[i].append(tempoptions)

    tempoptions = ""
    if (temp[n][:7] == "    b. "):
        tempoptions += temp[n][7:]
        n = n + 1
    while not temp[n].startswith("    c. "):
        tempoptions += "\n"
        tempoptions += temp[n][7:]
        n = n + 1
    options[i].append(tempoptions)

    tempoptions = ""
    if (temp[n][:7] == "    c. "):
        tempoptions += temp[n][7:]
        n = n + 1
    while not temp[n].startswith("    d. "):
        tempoptions += "\n"
        tempoptions += temp[n][7:]
        n = n + 1
    options[i].append(tempoptions)

    tempoptions = ""
    if (temp[n][:7] == "    d. "):
        tempoptions += temp[n][7:]
        n = n + 1
    while n < len(temp):
        tempoptions += "\n"
        tempoptions += temp[n][7:]
        n = n + 1
    options[i].append(tempoptions)

    data[str(i + 1)] = {
    "question": questions[i],
    "options": options[i]
    }

with open('list.json', 'w') as json_file:
    json.dump(data, json_file, indent=4)
```
